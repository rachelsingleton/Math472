## Question 1: Use the EM algorithm to fit a $(1-\pi)$ : $\pi$ mixture of two Poisson distributions, Poisson$(\lambda_1)$ and Poisson$(\lambda_2)$, to the given data.

### (1a) Derive the EM algorithm for the maximum likelihood estimates of $\pi,\lambda_1,\lambda_2$.

### Defining Important things for our derivation:

General Poisson distribution is Poisson$(\lambda) = \frac{e^{-\lambda}*\lambda^x}{x!}$

$f(Y) = f(x|z)f(z)$

$\theta = (\pi,\lambda_1,\lambda_2)^{T}$

General Likelihood function is $L(\theta | Y) = f(y;\theta) = f(x|z,\theta)f(z|\theta)$

### Derivation:

We define the pmf of random variable $X$ to be:

$f(x;\theta) = (1-\pi)Poi(x;\lambda_1) + \pi Poi(x;\lambda_2)$

which gives us a Likelihood function of:

$L(\theta|Y) = \prod_{i=1}^{n} [(1-z_i)Poi(x_i;\lambda_1)+z_iPoi(x_i;\lambda_2)]*[(1-\pi)^{1-z_i}\pi^{z_i}]$.

We can then take the log of this to get the log-likelihood function:

$log(L(\theta|Y)) = \sum_{i=1}^{n} [(1-z_i)log(Poi(x_i;\lambda_1))+z_ilog(Poi(x_i;\lambda_2))] + \sum_{i=1}^{n} [(1-z_i)log(1-\pi)+z_ilog(\pi)]$.

We can now define our $Q$ function as the expected value of the log-likelihood function which gives us

$Q(\theta|\theta^{(t)}) = E[log(L(\theta|Y))] = \sum_{i=1}^{n} [(1-z_i^{(t)})log(Poi(x_i;\lambda_1))+z_i^{(t)}log(Poi(x_i;\lambda_2))] + \sum_{i=1}^{n} [(1-z_i^{(t)})log(1-\pi)+z_i^{(t)}log(\pi)]$ where

$z_i^{(t)} = E[z_i|x,\theta^{(t)}] = P(z_i=1|z,\theta^{(t)}) = \frac{Poi(x_i;\lambda_2)\pi^{(t)}}{f(x_i;\theta)}$. This is our E-Step.

We will now take the partial derivatives of $Q$ so that we can solve for the MLEs of our unknown variables.

$\frac{\partial Q}{\partial \pi} = \sum_{i=1}^{n} [-\frac{1-z_i^{(t)}}{1-\pi} + \frac{z_i^{(t)}}{\pi}]$

$\frac{\partial Q}{\partial \lambda_1} = \sum_{i=1}^{n} [-(1-z_i^{(t)}) + \frac{(1-z_i^{(t)})x_i}{\lambda_1}]$

$\frac{\partial Q}{\partial \lambda_2} = \sum_{i=1}^{n} [-(z_i^{(t)}) + \frac{(z_i^{(t)})x_i}{\lambda_2}]$

We can now set all of these partial derivatives equal to $0$ and solve!

$-\sum_{i=1}^{n} \frac{1-z_i^{(t)}}{1-\pi} + \sum_{i=1}^{n} \frac{z_i^{(t)}}{\pi} = 0$

$\Rightarrow \sum_{i=1}^{n} \frac{z_i^{(t)}}{\pi} = \sum_{i=1}^{n} \frac{1-z_i^{(t)}}{1-\pi}$

$\Rightarrow \frac{1}{\pi} \sum_{i=1}^{n} z_i^{(t)} = \frac{1}{1-\pi} \sum_{i=1}^{n} 1-z_i^{(t)}$

$\Rightarrow (1-\pi) \sum_{i=1}^{n} z_i^{(t)} = \pi \sum_{i=1}^{n} 1-z_i^{(t)}$

$\Rightarrow \sum_{i=1}^{n} z_i^{(t)} - \pi \sum_{i=1}^{n} z_i^{(t)} = \pi \sum_{i=1}^{n} 1 - \pi \sum_{i=1}^{n} z_i^{(t)}$

$\Rightarrow \sum_{i=1}^{n} z_i^{(t)} = \pi n $

$\Rightarrow \hat\pi = \frac{\sum_{i=1}^{n} z_i^{(t)}}{n}$

------------------------------------------------------------------------------------------------------------------

$\sum_{i=1}^{n} [-(1-z_i^{(t)}) + \frac{(1-z_i^{(t)})x_i}{\lambda_1}] = 0$

$\Rightarrow -\sum_{i=1}^{n} [(1-z_i^{(t)})] + \sum_{i=1}^{n} [\frac{(1-z_i^{(t)})x_i}{\lambda_1}] = 0$

$\Rightarrow \sum_{i=1}^{n} [\frac{(1-z_i^{(t)})x_i}{\lambda_1}] = \sum_{i=1}^{n} [(1-z_i^{(t)})]$

$\Rightarrow \frac{1}{\lambda_1} \sum_{i=1}^{n} [(1-z_i^{(t)})x_i] = \sum_{i=1}^{n} 1 - \sum_{i=1}^{n} z_i^{(t)})$

$\Rightarrow \sum_{i=1}^{n} [(1-z_i^{(t)})x_i] = \lambda_1 (n - \sum_{i=1}^{n} z_i^{(t)}))$

$\Rightarrow \hat\lambda_1 = \frac{\sum_{i=1}^{n} [(1-z_i^{(t)})x_i]}{n - \sum_{i=1}^{n} (z_i^{(t)})}$

------------------------------------------------------------------------------------------------------------------

$\sum_{i=1}^{n} [-(z_i^{(t)}) + \frac{(z_i^{(t)})x_i}{\lambda_2}] = 0$

$\Rightarrow -\sum_{i=1}^{n} [(z_i^{(t)})] + \sum_{i=1}^{n} [\frac{(z_i^{(t)})x_i}{\lambda_2}] = 0$

$\Rightarrow \sum_{i=1}^{n} [\frac{(z_i^{(t)})x_i}{\lambda_2}] = \sum_{i=1}^{n} [(z_i^{(t)})]$

$\Rightarrow \frac{1}{\lambda_2} \sum_{i=1}^{n} [(z_i^{(t)})x_i] = \sum_{i=1}^{n} z_i^{(t)})$

$\Rightarrow \sum_{i=1}^{n} [(z_i^{(t)})x_i] = \lambda_2 \sum_{i=1}^{n} z_i^{(t)})$

$\Rightarrow \hat\lambda_2 = \frac{\sum_{i=1}^{n} [(z_i^{(t)})x_i]}{\sum_{i=1}^{n} z_i^{(t)}}$

The above answers give us our M-Step. 

### (1b) Implementing the EM algorithm for this data set. Present the output of your code in the format similar to Table 4.1 on page 102 of the textbook. 

```python
import numpy as np
from scipy.stats import poisson
import pandas as pd

zeros = 0*np.ones((162,1))
ones = np.ones((267,1))
twos = 2*np.ones((271,1))
threes = 3*np.ones((185,1))
fours = 4*np.ones((111,1))
fives = 5*np.ones((61,1))
sixes = 6*np.ones((27,1))
sevens = 7*np.ones((8,1))
eights = 8*np.ones((3,1))
nines = 9*np.ones((1,1))

data = np.vstack((zeros,ones,twos,threes,fours,fives,sixes,sevens,eights,nines))

pi = 1/2
pi_vals = [pi]
lamda_1 = 2
lamda1_vals = [lamda_1]
lamda_2 = 1
lamda2_vals = [lamda_2]


def E_step(pi,lamda_1,lamda_2):
    val = (poisson.pmf(data,lamda_2)*pi) / f(pi,lamda_1,lamda_2,data)
    return val
    
def f(pi,lamda_1,lamda_2,data):
    return ((1-pi)*(poisson.pmf(data, lamda_1))) + ((pi)*(poisson.pmf(data, lamda_2)))

def update_pi(pi,lamda_1,lamda_2):
    return np.sum(E_step(pi,lamda_1,lamda_2)) / 1096

def update_lam1(pi,lamda_1,lamda_2):
    lamda_1 = ((1-E_step(pi,lamda_1,lamda_2))*data) / (1096 - np.sum(E_step(pi,lamda_1,lamda_2)))
    return np.sum(lamda_1)
    
def update_lam2(pi,lamda_1,lamda_2):
    lamda_2 = ((E_step(pi,lamda_1,lamda_2))*data) / (np.sum(E_step(pi,lamda_1,lamda_2)))
    return np.sum(lamda_2)

def conv_criteria(pi,lamda_1,lamda_2,pi1,lamda_11,lamda_21):
    theta = np.array([pi,lamda_1,lamda_2]).reshape((3,1))
    theta1 = np.array([pi1,lamda_11,lamda_21]).reshape((3,1))
    r_top = np.linalg.norm(theta-theta1)
    r_bottom = np.linalg.norm(theta1)
    return r_top/r_bottom

values = []
appendlist = []
def EM_alg(pi,lamda_1,lamda_2):
    convergences = ['-']
    i = 0
    tolerance = 0.00001
    conv = 1
    max_iterations = 1000

    while(i < max_iterations):
        pi_val = update_pi(pi_vals[i],lamda1_vals[i],lamda2_vals[i])
        lam1_val = update_lam1(pi_vals[i],lamda1_vals[i],lamda2_vals[i])
        lam2_val = update_lam2(pi_vals[i],lamda1_vals[i],lamda2_vals[i])
                
        pi_vals.append(pi_val)
        lamda1_vals.append(lam1_val)
        lamda2_vals.append(lam2_val)
        
        convergences.append(conv_criteria(pi_val,lam1_val,lam2_val,pi_vals[i],lamda1_vals[i],lamda2_vals[i]))
        
#         DC = (pC-0.07083690804839426)/(pc[i]-0.07083690804839426)
#         ratiosDC.append(DC)
#         DI = (pI-0.1887367920620008)/(pi[i]-0.1887367920620008)
#         ratiosDI.append(DI)
        
        values = [i,pi_vals[i],lamda1_vals[i],lamda2_vals[i],convergences[i]]
        appendlist.append(values)
        i += 1
EM_alg(pi,lamda_1,lamda_2)

# print("Table for Starting Alphas [0.5,0.5] Using Newton's Method")
df = pd.DataFrame(appendlist, columns = ['Iteration,t','Pi','Lambda1','Lambda2','R^(t)'])
# df = df.iloc[2:7,:]
df = df.style.hide_index()
df = df.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
df
```

```python
from scipy.stats import multivariate_normal

data = pd.read_csv("data_mvnorm2mix.csv", delimiter=",", header = None)
data = np.array(data).reshape((800,2))

pi = 0.6
pi_vals = [pi]
mu_1 = np.array([1,0.1]).reshape((2,))
mu1_vals = [mu_1]
mu_2 = np.array([2,1]).reshape((2,))
mu2_vals = [mu_2]
sigma_1 = np.array([[2,0.4],[0.4,0.5]]).reshape((2,2))
sigma1_vals = [sigma_1]
sigma_2 = np.array([[1,-1], [-1,1.5]]).reshape((2,2))
sigma2_vals = [sigma_2]

def E_step(pi,mu_1,mu_2,sigma_1,sigma_2):
    val = (multivariate_normal.pdf(data,mu_2,sigma_2)*pi) / f(pi,mu_1,mu_2,sigma_1,sigma_2,data)
    return val
    
def f(pi,mu_1,mu_2,sigma_1,sigma_2,data):
    return ((1-pi)*(multivariate_normal.pdf(data,mu_1,sigma_1))) + ((pi)*(multivariate_normal.pdf(data,mu_2,sigma_2)))

def update_pi(z):
    val = np.sum(z) / 800
    return val

def update_mu1(z):
    zdata = (1-z)@data
    mu_1 = zdata / (800 - np.sum(z))
    return mu_1
    
def update_mu2(z):
    mu_2 = ((z)@data) / (np.sum(z))
    return mu_2

def update_sigma1(newmu_1,z):
    delta = data-newmu_1
    sigma_1 = ((1-z)*delta.T)@(delta) / (800 - np.sum(z))
    return sigma_1
               
def update_sigma2(newmu_2,z):
    delta = data-newmu_2
    sigma_2 = ((z)*delta.T).dot((delta)) / (np.sum(z))
    return sigma_2              
               
# def conv_criteria(pi,lamda_1,lamda_2,pi1,lamda_11,lamda_21):
#     theta = np.array([pi,lamda_1,lamda_2]).reshape((3,1))
#     theta1 = np.array([pi1,lamda_11,lamda_21]).reshape((3,1))
#     r_top = np.linalg.norm(theta-theta1)
#     r_bottom = np.linalg.norm(theta1)
#     return r_top/r_bottom

values = []
appendlist = []
def EM_alg(pi,mu_1,mu_2,sigma_1,sigma_2):
    convergences = ['-']
    i = 0
    max_iterations = 22

    while(i < max_iterations):
        z = E_step(pi_vals[-1],mu1_vals[-1],mu2_vals[-1],sigma1_vals[-1],sigma2_vals[-1])
        mu1_val = update_mu1(z)
        mu1_vals.append(mu1_val)
        mu2_val = update_mu2(z)
        mu2_vals.append(mu2_val)
        sigma1_val = update_sigma1(mu1_vals[-1],z)
        sigma2_val = update_sigma2(mu2_vals[-1],z)
                
        pi_val = update_pi(z)
        pi_vals.append(pi_val)
        sigma1_vals.append(sigma1_val)
        sigma2_vals.append(sigma2_val)
        
#         convergences.append(conv_criteria(pi_val,lam1_val,lam2_val,pi_vals[i],lamda1_vals[i],lamda2_vals[i]))
        
        values = [i,pi_vals[i],mu1_vals[i],mu2_vals[i],sigma1_vals[i],sigma2_vals[i]]
        appendlist.append(values)
        i += 1
EM_alg(pi,mu_1,mu_2,sigma_1,sigma_2)
               
# print("Table for Starting Alphas [0.5,0.5] Using Newton's Method")
df = pd.DataFrame(appendlist, columns = ['Iteration,t','Pi','Mu1','Mu2','Sigma1','Sigma2'])
df = df.style.hide_index()
df = df.set_table_styles([dict(selector='th', props=[('text-align', 'center')])])
df
```

### (2b) Plot the value of the log-likelihood function vs iteration number.  Comment on the pattern of the plot. 

```python
import matplotlib.pyplot as plt
%matplotlib inline

pi = 0.6
pi_vals = [pi]
mu_1 = np.array([1,0.1]).reshape((2,))
mu1_vals = [mu_1]
mu_2 = np.array([2,1]).reshape((2,))
mu2_vals = [mu_2]
sigma_1 = np.array([[2,0.4],[0.4,0.5]]).reshape((2,2))
sigma1_vals = [sigma_1]
sigma_2 = np.array([[1,-1], [-1,1.5]]).reshape((2,2))
sigma2_vals = [sigma_2]

def log_likelihood(pi,mu_1,mu_2,sigma_1,sigma_2):
    z = E_step(pi,mu_1,mu_2,sigma_1,sigma_2)
    last = np.sum((1-z)*np.log(1-pi) + z*np.log(pi))
    first = np.sum((1-z)*np.log(multivariate_normal.pdf(data,mu_1,sigma_1)) + z*np.log(multivariate_normal.pdf(data,mu_2,sigma_2)))
    return first + last

y_list = []
i_list = []

def EM_alg(pi,mu_1,mu_2,sigma_1,sigma_2):
    i = 0
    max_iterations = 50

    while(i < max_iterations):
        z = E_step(pi_vals[-1],mu1_vals[-1],mu2_vals[-1],sigma1_vals[-1],sigma2_vals[-1])
        y = log_likelihood(pi_vals[-1],mu1_vals[-1],mu2_vals[-1],sigma1_vals[-1],sigma2_vals[-1])
        y_list.append(y)
        i_list.append(i)
        mu1_val = update_mu1(z)
        mu1_vals.append(mu1_val)
        mu2_val = update_mu2(z)
        mu2_vals.append(mu2_val)
        sigma1_val = update_sigma1(mu1_vals[-1],z)
        sigma2_val = update_sigma2(mu2_vals[-1],z)
                
        pi_val = update_pi(z)
        pi_vals.append(pi_val)
        sigma1_vals.append(sigma1_val)
        sigma2_vals.append(sigma2_val)
        
        values = [i,pi_vals[i],mu1_vals[i],mu2_vals[i],sigma1_vals[i],sigma2_vals[i]]
        appendlist.append(values)
        i += 1
EM_alg(pi,mu_1,mu_2,sigma_1,sigma_2)
plt.plot(i_list,y_list)
plt.xlabel('Iteration',fontsize=20)
plt.ylabel('Log-Likelihood',fontsize=20)
plt.title('Iteration vs Log-Likelihood',fontsize=17)
plt.show()
```
